````python
# ── 재현성 고정 (팀 공통) ───────────────────────────────────────
def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed_everything(42)

DATA_DIR = os.path.join(os.path.expanduser("~"), "work/semantic_segmentation/data/training")
MODEL_DIR = os.path.join(os.path.expanduser("~"), "work/semantic_segmentation")
os.makedirs(MODEL_DIR, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device : {DEVICE}")
print(f"DATA_DIR: {DATA_DIR}")
````

````python
# ── Augmentation (팀 공통 고정값) ──────────────────────────────
def build_augmentation(is_train=True):
    """두 모델에 완전히 동일한 augmentation 적용"""
    if is_train:
        return Compose([
            HorizontalFlip(p=0.5),
            RandomSizedCrop(
                min_max_height=(300, 370),
                w2h_ratio=370/1242,
                height=224, width=224,
                size=(224, 224),
                p=0.5
            ),
            Resize(width=224, height=224)
        ])
    return Compose([Resize(width=224, height=224)])
````

````python
# Augmentation 시각화
aug_test  = build_augmentation(is_train=False)
aug_train = build_augmentation(is_train=True)

input_images = sorted(glob(os.path.join(DATA_DIR, "image_2", "*.png")))

plt.figure(figsize=(12, 20))
for i in range(5):
    image = imread(input_images[i])
    resized   = aug_test( image=image)["image"]
    processed = aug_train(image=image)["image"]
    plt.subplot(5, 2, 2*i+1)
    plt.imshow(resized);  plt.title("Original (resized)"); plt.axis('off')
    plt.subplot(5, 2, 2*i+2)
    plt.imshow(processed); plt.title("Augmented");         plt.axis('off')
plt.suptitle("Augmentation 예시 (왼쪽: 원본 resize / 오른쪽: augmented)", fontsize=13)
plt.tight_layout()
plt.savefig(os.path.join(MODEL_DIR, "augmentation_sample.png"), dpi=100, bbox_inches='tight')
plt.show()
print("저장 완료: augmentation_sample.png")
````

````python
# ── 데이터 분할 고정 (팀 공통 CSV) ────────────────────────────
SPLIT_FILE = os.path.join(MODEL_DIR, "split.csv")

def make_fixed_split(data_dir, split_file, test_n=30):
    """Train/Test 분할을 CSV로 저장 (seed=42 고정)"""
    if os.path.exists(split_file):
        print(f"기존 split 파일 사용: {split_file}")
        return
    imgs = sorted(glob(os.path.join(data_dir, "image_2",  "*.png")))
    lbls = sorted(glob(os.path.join(data_dir, "semantic", "*.png")))
    assert len(imgs) == len(lbls), f"이미지/라벨 수 불일치: {len(imgs)} vs {len(lbls)}"
    pairs = list(zip(imgs, lbls))
    random.shuffle(pairs)   # seed=42 로 고정됨
    with open(split_file, "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["img", "lbl", "split"])
        for img, lbl in pairs[:-test_n]: w.writerow([img, lbl, "train"])
        for img, lbl in pairs[-test_n:]: w.writerow([img, lbl, "test"])
    print(f"Split 저장 완료: {len(pairs)-test_n} train / {test_n} test")

make_fixed_split(DATA_DIR, SPLIT_FILE)
````

````python
class KittiDataset(Dataset):
    """
    KITTI Semantic Segmentation Dataset
    - 라벨값 7 = 도로(road) → 이진 마스크(0/1) 변환
    """
    def __init__(self, split="train", augmentation=None):
        self.augmentation = augmentation
        self.data = []
        with open(SPLIT_FILE) as f:
            for row in csv.DictReader(f):
                if row["split"] == split:
                    self.data.append((row["img"], row["lbl"]))

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        img_path, lbl_path = self.data[idx]
        _input  = imread(img_path)
        _output = imread(lbl_path)
        _output = (_output == 7).astype(np.uint8)   # road=1, else=0

        data = {"image": _input, "mask": _output}
        if self.augmentation:
            aug     = self.augmentation(**data)
            _input  = aug["image"] / 255.0
            _output = aug["mask"]

        _output = np.expand_dims(_output, 0)   # (H,W) → (1,H,W)
        return (
            torch.tensor(_input,  dtype=torch.float32).permute(2, 0, 1),  # (C,H,W)
            torch.tensor(_output, dtype=torch.float32)                     # (1,H,W)
        )

# DataLoader (UNet++ 메모리 고려 → batch_size=8)
train_dataset = KittiDataset("train", augmentation=build_augmentation(True))
test_dataset  = KittiDataset("test",  augmentation=build_augmentation(False))

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True,  num_workers=0)
test_loader  = DataLoader(test_dataset,  batch_size=8, shuffle=False, num_workers=0)

print(f"Train: {len(train_dataset)} / Test: {len(test_dataset)}")
x, y = next(iter(train_loader))
print(f"Input shape : {x.shape}  |  Mask shape: {y.shape}")
````

## 2. 모델 구조

### 2-1. U-Net (기준 모델)

```
Input(3,224,224)
  ↓ Encoder: Conv×2 + MaxPool  ×4
  ↓ Bottleneck + Dropout(0.5)
  ↓ Decoder: ConvTranspose + Skip(cat) + Conv×2  ×4
  ↓ Conv(1×1) + Sigmoid
Output(1,224,224)
```

### 2-2. U-Net++ (Dense Skip Connection)

기존 U-Net의 단순 skip connection을 **nested dense skip path**로 교체:

$$x^{i,j} = \mathcal{H}\left([x^{i,k}]_{k=0}^{j-1},\ \mathcal{U}(x^{i+1,j-1})\right)$$

- $x^{i,j}$: row $i$, column $j$의 feature map
- $\mathcal{U}$: 2×2 ConvTranspose 업샘플링
- $[\cdot]$: channel-wise concat

**[Ablation] base_channels 변화:**

| | ch=8 | ch=16 | ch=32 |
|--|-----|-------|-------|
| filters | [8,16,32,64] | [16,32,64,128] | [32,64,128,256] |
| 파라미터 | 소형 | 중형 | 대형 |

````python
# ── 공통 블록 ──────────────────────────────────────────────────
class DoubleConv(nn.Module):
    """Conv(3×3) → BN → ReLU → Conv(3×3) → BN → ReLU"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(in_channels,  out_channels, 3, padding=1),
            nn.BatchNorm2d(out_channels), nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, 3, padding=1),
            nn.BatchNorm2d(out_channels), nn.ReLU(inplace=True),
        )
    def forward(self, x): return self.net(x)
````

````python
class UNet(nn.Module):
    """
    U-Net 기준 모델
    base_channels: 첫 인코더 블록 채널 수 (기본 64)
    """
    def __init__(self, input_channels=3, output_channels=1, base_channels=64):
        super().__init__()
        b = base_channels

        # Encoder
        self.enc1 = DoubleConv(input_channels, b);     self.pool1 = nn.MaxPool2d(2)
        self.enc2 = DoubleConv(b,   b*2);              self.pool2 = nn.MaxPool2d(2)
        self.enc3 = DoubleConv(b*2, b*4);              self.pool3 = nn.MaxPool2d(2)
        self.enc4 = DoubleConv(b*4, b*8);              self.pool4 = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = DoubleConv(b*8, b*16)
        self.dropout    = nn.Dropout(0.5)

        # Decoder
        self.up6  = nn.ConvTranspose2d(b*16, b*8, 2, 2);  self.dec6 = DoubleConv(b*16, b*8)
        self.up7  = nn.ConvTranspose2d(b*8,  b*4, 2, 2);  self.dec7 = DoubleConv(b*8,  b*4)
        self.up8  = nn.ConvTranspose2d(b*4,  b*2, 2, 2);  self.dec8 = DoubleConv(b*4,  b*2)
        self.up9  = nn.ConvTranspose2d(b*2,  b,   2, 2);  self.dec9 = DoubleConv(b*2,  b)

        self.final = nn.Conv2d(b, output_channels, 1)

    def forward(self, x):
        c1 = self.enc1(x);          p1 = self.pool1(c1)
        c2 = self.enc2(p1);         p2 = self.pool2(c2)
        c3 = self.enc3(p2);         p3 = self.pool3(c3)
        c4 = self.enc4(p3);         p4 = self.pool4(c4)
        c5 = self.dropout(self.bottleneck(p4))

        u6 = self.up6(c5);  c6 = self.dec6(torch.cat([u6, c4], 1))
        u7 = self.up7(c6);  c7 = self.dec7(torch.cat([u7, c3], 1))
        u8 = self.up8(c7);  c8 = self.dec8(torch.cat([u8, c2], 1))
        u9 = self.up9(c8);  c9 = self.dec9(torch.cat([u9, c1], 1))
        return torch.sigmoid(self.final(c9))

# 파라미터 수 확인
for b in [32, 64]:
    m = UNet(base_channels=b)
    p = sum(x.numel() for x in m.parameters() if x.requires_grad)
    print(f"UNet  base={b:3d}: {p:>12,} params")
````

````python
class UNetPlusPlus(nn.Module):
    """
    U-Net++ (Nested U-Net)
    Zhou et al., 2018 (v1), 2020 (v2)

    핵심: x^{i,j} = H( [x^{i,0..j-1}], U(x^{i+1,j-1}) )
    dense skip path로 인코더-디코더 간 semantic gap 감소

    [Ablation] base_channels: 8 / 16 / 32
    deep_supervision=False: 최종 출력(x^{0,4})만 사용
    """
    def __init__(self, input_channels=3, output_channels=1,
                 base_channels=32, deep_supervision=False):
        super().__init__()
        self.deep_supervision = deep_supervision
        nb = base_channels
        f  = [nb, nb*2, nb*4, nb*8, nb*16]   # 채널 수 리스트

        # ── Encoder X^{i,0} ──────────────────────────────────
        self.enc = nn.ModuleList([
            DoubleConv(input_channels, f[0]),
            DoubleConv(f[0], f[1]),
            DoubleConv(f[1], f[2]),
            DoubleConv(f[2], f[3]),
            DoubleConv(f[3], f[4]),   # bottleneck
        ])
        self.pool = nn.ModuleList([nn.MaxPool2d(2) for _ in range(4)])

        # ── Dense nodes & Upsamplers ──────────────────────────
        self.up = nn.ModuleDict()
        self.nd = nn.ModuleDict()

        # j=1: X^{i,1}, i=0..3
        for i in range(4):
            self.up[f"{i}_1"] = nn.ConvTranspose2d(f[i+1], f[i+1], 2, 2)
            self.nd[f"{i}_1"] = DoubleConv(f[i]*1 + f[i+1], f[i])
        # j=2: X^{i,2}, i=0..2
        for i in range(3):
            self.up[f"{i}_2"] = nn.ConvTranspose2d(f[i+1], f[i+1], 2, 2)
            self.nd[f"{i}_2"] = DoubleConv(f[i]*2 + f[i+1], f[i])
        # j=3: X^{i,3}, i=0..1
        for i in range(2):
            self.up[f"{i}_3"] = nn.ConvTranspose2d(f[i+1], f[i+1], 2, 2)
            self.nd[f"{i}_3"] = DoubleConv(f[i]*3 + f[i+1], f[i])
        # j=4: X^{0,4}
        self.up["0_4"] = nn.ConvTranspose2d(f[1], f[1], 2, 2)
        self.nd["0_4"] = DoubleConv(f[0]*4 + f[1], f[0])

        # ── Output heads (deep supervision 시 4개) ───────────
        self.out_heads = nn.ModuleList([
            nn.Conv2d(f[0], output_channels, 1) for _ in range(4)
        ])

    def forward(self, x):
        # Encoder
        e = []; cur = x
        for i in range(4):
            cur = self.enc[i](cur); e.append(cur); cur = self.pool[i](cur)
        e.append(self.enc[4](cur))          # x^{4,0} = bottleneck

        # n[i][j] = X^{i,j}
        n = [[None]*5 for _ in range(5)]
        for i in range(5): n[i][0] = e[i]

        # j=1
        for i in range(4):
            up = self.up[f"{i}_1"](n[i+1][0])
            n[i][1] = self.nd[f"{i}_1"](torch.cat([n[i][0], up], 1))
        # j=2
        for i in range(3):
            up = self.up[f"{i}_2"](n[i+1][1])
            n[i][2] = self.nd[f"{i}_2"](torch.cat([n[i][0], n[i][1], up], 1))
        # j=3
        for i in range(2):
            up = self.up[f"{i}_3"](n[i+1][2])
            n[i][3] = self.nd[f"{i}_3"](torch.cat([n[i][0], n[i][1], n[i][2], up], 1))
        # j=4
        up = self.up["0_4"](n[1][3])
        n[0][4] = self.nd["0_4"](torch.cat([n[0][0], n[0][1], n[0][2], n[0][3], up], 1))

        outs = [torch.sigmoid(self.out_heads[j](n[0][j+1])) for j in range(4)]
        if self.deep_supervision:
            return outs          # [x^{0,1}, x^{0,2}, x^{0,3}, x^{0,4}]
        return outs[-1]          # x^{0,4} 최종 출력만

# 파라미터 수 확인
for b in [8, 16, 32]:
    m = UNetPlusPlus(base_channels=b)
    p = sum(x.numel() for x in m.parameters() if x.requires_grad)
    print(f"UNet++ base={b:3d}: {p:>12,} params")
````

## 3. 손실 함수 : BCE + Dice Loss

U-Net++ 논문에서 제안하는 손실함수:
$$\mathcal{L} = \mathcal{L}_{BCE} + (1 - \text{Dice})$$

$$\text{Dice} = \frac{2 \cdot |P \cap T| + \epsilon}{|P| + |T| + \epsilon}$$

````python
def dice_coef(pred, target, smooth=1.0):
    """Dice Coefficient (미분 가능한 버전)"""
    pred_f   = pred.view(-1)
    target_f = target.view(-1)
    intersection = (pred_f * target_f).sum()
    return (2. * intersection + smooth) / (pred_f.sum() + target_f.sum() + smooth)

def bce_dice_loss(pred, target):
    """BCE + Dice Loss (U-Net++ 논문 기준)"""
    bce  = F.binary_cross_entropy(pred, target)
    dice = 1.0 - dice_coef(pred, target)
    return bce + dice

# 동작 확인
dummy_pred = torch.sigmoid(torch.randn(2, 1, 224, 224))
dummy_tgt  = torch.randint(0, 2, (2, 1, 224, 224)).float()
print(f"BCE+Dice Loss (random): {bce_dice_loss(dummy_pred, dummy_tgt):.4f}")
````

## 4. 학습 / 평가 함수

````python
def train_one_epoch(model, loader, optimizer):
    model.train()
    total_loss = 0.0
    for imgs, masks in loader:
        imgs, masks = imgs.to(DEVICE), masks.to(DEVICE)
        optimizer.zero_grad()
        out  = model(imgs)
        pred = out[-1] if isinstance(out, list) else out
        loss = bce_dice_loss(pred, masks)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

def evaluate(model, loader):
    model.eval()
    total_loss = 0.0
    with torch.no_grad():
        for imgs, masks in loader:
            imgs, masks = imgs.to(DEVICE), masks.to(DEVICE)
            out  = model(imgs)
            pred = out[-1] if isinstance(out, list) else out
            total_loss += bce_dice_loss(pred, masks).item()
    return total_loss / len(loader)

def compute_miou(model, dataset, threshold=0.5):
    """도로(road) 클래스 IoU 계산 (mIoU = road IoU만 계산)"""
    model.eval()
    ious = []
    loader = DataLoader(dataset, batch_size=8, shuffle=False)
    with torch.no_grad():
        for imgs, masks in loader:
            imgs = imgs.to(DEVICE)
            out  = model(imgs)
            preds = out[-1] if isinstance(out, list) else out
            preds = (preds.cpu().numpy() > threshold).astype(np.uint8)
            masks = masks.numpy().astype(np.uint8)
            for p, m in zip(preds, masks):
                inter = np.logical_and(p, m).sum()
                union = np.logical_or(p, m).sum()
                ious.append(inter / (union + 1e-7))
    return float(np.mean(ious))

def run_experiment(model, train_loader, test_loader, num_epochs=30, lr=1e-4):
    """모델 학습 + 기록 반환"""
    model = model.to(DEVICE)
    optimizer = optim.Adam(model.parameters(), lr=lr)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

    train_losses, val_losses = [], []
    for epoch in range(1, num_epochs + 1):
        tr_loss = train_one_epoch(model, train_loader, optimizer)
        vl_loss = evaluate(model, test_loader)
        scheduler.step()
        train_losses.append(tr_loss)
        val_losses.append(vl_loss)
        if epoch % 5 == 0:
            print(f"  Epoch {epoch:3d}/{num_epochs} | Train: {tr_loss:.4f} | Val: {vl_loss:.4f}")
    return train_losses, val_losses

print("학습 함수 정의 완료")
````

## 5. 실험 실행

### Step 1. U-Net 기준 모델 학습

````python
%%time
print("=" * 55)
print("  U-Net (base_channels=64) 학습")
print("=" * 55)

model_unet = UNet(input_channels=3, output_channels=1, base_channels=64)
tr_unet, vl_unet = run_experiment(
    model_unet, train_loader, test_loader, num_epochs=30, lr=1e-4
)
iou_unet = compute_miou(model_unet, test_dataset)
params_unet = sum(p.numel() for p in model_unet.parameters() if p.requires_grad)

print(f"\n  ✅ U-Net  IoU : {iou_unet:.4f}")
print(f"     파라미터    : {params_unet:,}")
torch.save(model_unet.state_dict(), os.path.join(MODEL_DIR, "UNet_base64.pth"))
````

### Step 2. U-Net++ 채널 차원수 Ablation Study

세 가지 `base_channels` (8, 16, 32) 를 동일 조건에서 비교합니다.

````python
%%time
# 채널 차원수 실험 목록
channel_configs = [8, 16, 32]
unetpp_results  = {}

for base_ch in channel_configs:
    name = f"UNetPP_ch{base_ch}"
    print("\n" + "="*55)
    print(f"  U-Net++ (base_channels={base_ch}) 학습")
    print("="*55)

    seed_everything(42)   # 각 실험마다 시드 재고정
    model = UNetPlusPlus(
        input_channels=3, output_channels=1,
        base_channels=base_ch, deep_supervision=False
    )
    params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"  파라미터 수: {params:,}")

    tr_ls, vl_ls = run_experiment(
        model, train_loader, test_loader, num_epochs=30, lr=1e-4
    )
    iou = compute_miou(model, test_dataset)

    unetpp_results[name] = {
        "model": model, "base_ch": base_ch,
        "train": tr_ls, "val": vl_ls,
        "iou": iou, "params": params
    }
    print(f"  ✅ {name}  IoU: {iou:.4f}")
    torch.save(model.state_dict(), os.path.join(MODEL_DIR, f"{name}.pth"))
````

## 6. 정량적 비교 (IoU)

## 7. 정성적 비교 (Segmentation Overlay)

````python
def calculate_iou_score(target, prediction):
    """단일 이미지 IoU 계산"""
    if target.shape != prediction.shape:
        prediction = resize(prediction, target.shape,
                            mode="constant", preserve_range=True).astype(np.uint8)
    inter = np.logical_and(target, prediction).sum()
    union = np.logical_or(target, prediction).sum()
    return float(inter / (union + 1e-7))

def get_output(model, image_path, label_path=None):
    """
    단일 이미지에 대해 추론 + 오버레이 시각화
    Returns: blended(PIL), prediction(np.uint8 binary), target(np.uint8 binary|None)
    """
    origin = imread(image_path)
    aug    = build_augmentation(is_train=False)
    proc   = aug(image=origin)["image"]

    tensor = (torch.tensor(proc / 255.0, dtype=torch.float32)
              .permute(2, 0, 1).unsqueeze(0).to(DEVICE))

    model.eval()
    with torch.no_grad():
        out  = model(tensor)
        pred = out[-1] if isinstance(out, list) else out

    pred_np = (pred[0].squeeze().cpu().numpy() > 0.5).astype(np.uint8)
    pred_resized = resize(pred_np, (origin.shape[0], origin.shape[1]),
                          mode="constant", preserve_range=True).astype(np.uint8)

    # 오버레이
    mask_img = Image.fromarray((pred_resized * 255).astype(np.uint8)).convert("L")
    background = Image.fromarray(origin).convert("RGBA")
    mask_rgba  = mask_img.resize((origin.shape[1], origin.shape[0])).convert("RGBA")
    blended    = Image.blend(background, mask_rgba, alpha=0.5)

    target = None
    if label_path:
        lbl      = imread(label_path)
        lbl_proc = aug(image=lbl)["image"]
        tgt_small = (lbl_proc == 7).astype(np.uint8)
        target    = resize(tgt_small, (origin.shape[0], origin.shape[1]),
                           mode="constant", preserve_range=True).astype(np.uint8)
    return blended, pred_resized, target

print("시각화 함수 정의 완료")
````

````python
# ── 단일 이미지 추론 & IoU 확인 ────────────────────────────────
test_imgs = sorted(glob(os.path.join(DATA_DIR, "image_2",  "*.png")))[-5:]
test_lbls = sorted(glob(os.path.join(DATA_DIR, "semantic", "*.png")))[-5:]

i = 0   # 확인할 이미지 인덱스 (0~4 변경 가능)
img_path = test_imgs[i]
lbl_path = test_lbls[i]

blended, pred, target = get_output(model_unet, img_path, lbl_path)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].imshow(imread(img_path));  axes[0].set_title("원본 이미지");  axes[0].axis("off")
axes[1].imshow(target, cmap="gray"); axes[1].set_title("정답 마스크 (road=1)"); axes[1].axis("off")
axes[2].imshow(np.array(blended)); axes[2].set_title(f"U-Net 예측 Overlay\nIoU={calculate_iou_score(target, pred):.4f}"); axes[2].axis("off")
plt.tight_layout(); plt.show()
````

````python
# ── 전체 모델 비교 시각화 (3장의 테스트 이미지) ───────────────
n_imgs   = 3
n_models = len(all_results)

fig, axes = plt.subplots(n_imgs, n_models + 1,
                          figsize=(4 * (n_models + 1), 4 * n_imgs))

for row, (img_p, lbl_p) in enumerate(zip(test_imgs[:n_imgs], test_lbls[:n_imgs])):
    origin = imread(img_p)
    axes[row, 0].imshow(origin)
    axes[row, 0].set_title("원본" if row == 0 else "", fontsize=11)
    axes[row, 0].axis("off")

    for col, (name, r) in enumerate(all_results.items()):
        blended, pred, target = get_output(r["model"], img_p, lbl_p)
        iou = calculate_iou_score(target, pred) if target is not None else 0.0
        axes[row, col + 1].imshow(np.array(blended))
        axes[row, col + 1].set_title(
            f"{name}\nIoU={iou:.4f}" if row == 0 else f"IoU={iou:.4f}",
            fontsize=9
        )
        axes[row, col + 1].axis("off")

plt.suptitle("Segmentation Overlay 비교 (테스트 이미지 3장)",
             fontsize=14, fontweight="bold", y=1.01)
plt.tight_layout()
plt.savefig(os.path.join(MODEL_DIR, "segmentation_overlay.png"),
            dpi=120, bbox_inches="tight")
plt.show()
print("저장: segmentation_overlay.png")
````

## 9. 결론 및 분석

### 채널 차원수 Ablation 결론

| 모델 | base_ch | IoU | 파라미터 | 특징 |
|------|---------|-----|---------|------|
| **U-Net** | 64 | **0.7950** | 31,043,521 | 빠른 수렴, 안정적 기준선 |
| U-Net++ ch=8  |  8 | 0.7690 | 690,060 | 채널 부족 → 표현력 한계, U-Net 대비 -3.3% |
| U-Net++ ch=16 | 16 | 0.7903 | 2,753,428 | U-Net에 근접, 파라미터 효율 최고 |
| **U-Net++ ch=32** | 32 | **0.8096** | 11,000,100 | **최고 IoU, U-Net 대비 +1.8% ★** |

---

### 정량 결과 시각화 해석

```
IoU
0.81 │                                    ★ UNet++ ch=32 (0.8096)
0.80 │       ● UNet base64 (0.7950)
0.79 │                         ◆ UNet++ ch=16 (0.7903)
0.77 │   ▲ UNet++ ch=8 (0.7690)
     └──────────────────────────────────────▶ base_channels
          8          16          32          64
```

---

### 핵심 발견

**① 채널 수 부족 (ch=8, IoU=0.7690 ↓)**  
U-Net 대비 IoU **-3.3%** 하락. U-Net++의 dense skip path가 표현력을 발휘하려면 각 node에 충분한 채널이 필요한데, ch=8에서는 concat 후에도 feature map이 너무 얇아 정보 손실이 발생. 파라미터(690K)는 가장 작지만 성능도 가장 낮아 **비효율적인 구간**.

**② 적정 채널 (ch=16, IoU=0.7903 ↔)**  
U-Net 대비 IoU **-0.6%** (거의 동등). 파라미터는 U-Net의 **8.9%** 수준(2.75M)으로 **파라미터 효율성이 가장 높은 구간**. 경량화가 필요한 실제 배포 환경에서 최적 선택.

**③ 넓은 채널 (ch=32, IoU=0.8096 ↑)**  
U-Net 대비 IoU **+1.8% 향상**. dense node들이 충분한 채널(32→64→128→256)을 가져 미세한 도로 경계 표현이 가능. 파라미터(11M)는 U-Net의 35% 수준으로 **성능과 파라미터 균형이 좋음**.  
→ **U-Net++ 이론적 장점이 실제로 확인된 구간**.

**④ U-Net 기준(ch=64) vs U-Net++**  
- U-Net은 단순 구조 덕에 동일 에폭에서 더 빠르게 수렴
- U-Net++는 `base_channels ≥ 32` 이상에서 U-Net을 추월
- dense skip connection의 이점은 충분한 채널 폭이 전제될 때 발현됨

---

### 파라미터 효율성 분석

| 모델 | IoU | Params | IoU/Params (×10⁶) |
|------|-----|--------|--------------------|
| U-Net base64 | 0.7950 | 31.0M | 0.0256 |
| UNet++ ch=8  | 0.7690 |  0.7M | **1.114** |
| UNet++ ch=16 | 0.7903 |  2.8M | 0.287 |
| UNet++ ch=32 | 0.8096 | 11.0M | 0.074 |

> ch=8은 파라미터 대비 IoU 비율이 가장 높지만, 절대 성능이 낮아 실용적이지 않음.  
> **실용적 추천: ch=16** (U-Net 동등 성능, 1/11 파라미터) 또는 **ch=32** (최고 성능)

---

### 팀 협업 관점

| 담당자 | 실험 | 주요 발견 |
|--------|------|----------|
| 배성렬 | 층수 조절 (L3/L5) | 깊을수록 항상 좋지 않음, KITTI에선 4층이 최적 |
| 최병훈 | Deep Supervision | 중간 출력 loss → 학습 초반 수렴 가속, 경계선 정교화 |
| **이희선** | **채널 차원수** | **ch=16~32 구간이 U-Net++ 실용 최적 구간** |
| 최수정 | width | 채널 차원수와 상호작용 추가 분석 필요 |

---
