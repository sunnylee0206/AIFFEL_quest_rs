## 4. 데이터로더 구성

**처리 순서**: 중복제거 → NaN제거 → MeCab 토크나이징 → 불용어 제거 → 사전 구성 → 인덱스 변환

```python
from konlpy.tag import Mecab
import numpy as np
from collections import Counter

tokenizer = Mecab()

stopwords = [
    '의','가','이','은','들','는','좀','잘','걍','과','도','를',
    '으로','자','에','와','한','하다','그','저','것','수','이다',
    '있다','하다','되다','않다','없다','나','우리','저','거'
]


def load_data(train_df, test_df, num_words=10000):
    """NSMC 데이터를 전처리하여 인덱스 시퀀스로 변환합니다."""

    # ① 중복 제거 + NaN 결측치 제거
    train_df = train_df.drop_duplicates(subset=['document']).dropna(how='any').copy()
    test_df  = test_df.drop_duplicates(subset=['document']).dropna(how='any').copy()
    print(f'전처리 후 Train: {len(train_df):,}건  |  Test: {len(test_df):,}건')

    # ② MeCab 토크나이징 + 불용어 제거
    def tokenize_sentences(sentences):
        result = []
        for sent in sentences:
            tokens = tokenizer.morphs(str(sent))
            tokens = [w for w in tokens if w not in stopwords and len(w) > 1]
            result.append(tokens)
        return result

    print('Train 토크나이징 중...')
    X_train = tokenize_sentences(train_df['document'])
    print('Test  토크나이징 중...')
    X_test  = tokenize_sentences(test_df['document'])

    # ③ 사전 구성 (빈도 상위 num_words, 특수 토큰 포함)
    all_words = np.concatenate(X_train).tolist()
    counter   = Counter(all_words).most_common(num_words - 4)
    vocab     = ['<PAD>', '<BOS>', '<EOS>', '<UNK>'] + [w for w, _ in counter]
    word_to_index = {w: i for i, w in enumerate(vocab)}
    print(f'사전 크기: {len(word_to_index):,}')

    # ④ 단어 리스트 → 인덱스 리스트 변환
    def encode(wordlist):
        return [word_to_index.get(w, word_to_index['<UNK>']) for w in wordlist]

    X_train = list(map(encode, X_train))
    X_test  = list(map(encode, X_test))

    return (X_train, np.array(train_df['label']),
            X_test,  np.array(test_df['label']),
            word_to_index)


X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
index_to_word = {v: k for k, v in word_to_index.items()}
```

```python
# 인코딩 / 디코딩 유틸 함수
def get_encoded_sentence(sentence, word_to_index):
    """문장 문자열을 <BOS> 포함 인덱스 리스트로 변환"""
    return ([word_to_index['<BOS>']] +
            [word_to_index.get(w, word_to_index['<UNK>'])
             for w in tokenizer.morphs(sentence)
             if w not in stopwords and len(w) > 1])

def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(s, word_to_index) for s in sentences]

def get_decoded_sentence(encoded, index_to_word):
    """인덱스 리스트를 문자열로 복원 (<BOS> 제외)"""
    return ' '.join(index_to_word.get(i, '<UNK>') for i in encoded[1:])

def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(e, index_to_word) for e in encoded_sentences]

# 샘플 확인
sample_sent = train_data['document'].dropna().iloc[1]
enc = get_encoded_sentence(sample_sent, word_to_index)
print(f'원문   : {sample_sent}')
print(f'인코딩 : {enc[:10]}...')
print(f'디코딩 : {get_decoded_sentence(enc, index_to_word)}')
```

```python
# MAX_LEN 결정 및 패딩/자르기 처리
MAX_LEN = 30   # 95퍼센타일 기준 설정

def pad_sequences(sequences, max_len, pad_idx=0):
    result = []
    for seq in sequences:
        seq = seq[:max_len]
        seq = seq + [pad_idx] * (max_len - len(seq))
        result.append(seq)
    return np.array(result, dtype=np.int64)

X_train_pad = pad_sequences(X_train, MAX_LEN)
X_test_pad  = pad_sequences(X_test,  MAX_LEN)

print(f'X_train_pad : {X_train_pad.shape}')
print(f'X_test_pad  : {X_test_pad.shape}')
```

## 6. 모델 구성

다음 **4가지 모델**을 실험합니다:
1. **SimpleRNN** — 기본 베이스라인
2. **LSTM** — 장기 의존성 처리
3. **Bidirectional LSTM** — 양방향 문맥 활용
4. **TextCNN** — 다중 커널 크기 병렬 컨볼루션

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split

DEVICE     = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
VOCAB_SIZE = len(word_to_index)
EMB_DIM    = 128
HIDDEN_DIM = 256
N_LAYERS   = 2
DROPOUT    = 0.3
BATCH_SIZE = 256
EPOCHS     = 10

print('Device:', DEVICE)

# 텐서 변환
X_tr = torch.tensor(X_train_pad, dtype=torch.long)
y_tr = torch.tensor(y_train,     dtype=torch.float)
X_te = torch.tensor(X_test_pad,  dtype=torch.long)
y_te = torch.tensor(y_test,      dtype=torch.float)

# Train / Validation 분리 (9:1)
full_ds  = TensorDataset(X_tr, y_tr)
val_size = int(len(full_ds) * 0.1)
tr_size  = len(full_ds) - val_size
train_ds, val_ds = random_split(full_ds, [tr_size, val_size],
                                generator=torch.Generator().manual_seed(42))
test_ds = TensorDataset(X_te, y_te)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  num_workers=2)
val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, num_workers=2)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, num_workers=2)

print(f'Train: {tr_size:,}  |  Val: {val_size:,}  |  Test: {len(test_ds):,}')
```

```python
# ──────────────────────────────────────────────────────────────
# 모델 1: SimpleRNN
# ──────────────────────────────────────────────────────────────
class SimpleRNNModel(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim, n_layers, dropout,
                 pretrained_emb=None, freeze_emb=False):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        if pretrained_emb is not None:
            self.embedding.weight.data.copy_(pretrained_emb)
        if freeze_emb:
            self.embedding.weight.requires_grad = False
        self.rnn     = nn.RNN(emb_dim, hidden_dim, n_layers, batch_first=True,
                              dropout=dropout if n_layers > 1 else 0)
        self.fc      = nn.Linear(hidden_dim, 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        emb = self.dropout(self.embedding(x))
        _, hidden = self.rnn(emb)
        return self.fc(self.dropout(hidden[-1])).squeeze(1)


# ──────────────────────────────────────────────────────────────
# 모델 2: LSTM
# ──────────────────────────────────────────────────────────────
class LSTMModel(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim, n_layers, dropout,
                 pretrained_emb=None, freeze_emb=False):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        if pretrained_emb is not None:
            self.embedding.weight.data.copy_(pretrained_emb)
        if freeze_emb:
            self.embedding.weight.requires_grad = False
        self.lstm    = nn.LSTM(emb_dim, hidden_dim, n_layers, batch_first=True,
                               dropout=dropout if n_layers > 1 else 0)
        self.fc      = nn.Linear(hidden_dim, 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        emb = self.dropout(self.embedding(x))
        _, (hidden, _) = self.lstm(emb)
        return self.fc(self.dropout(hidden[-1])).squeeze(1)


# ──────────────────────────────────────────────────────────────
# 모델 3: Bidirectional LSTM
# ──────────────────────────────────────────────────────────────
class BiLSTMModel(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim, n_layers, dropout,
                 pretrained_emb=None, freeze_emb=False):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        if pretrained_emb is not None:
            self.embedding.weight.data.copy_(pretrained_emb)
        if freeze_emb:
            self.embedding.weight.requires_grad = False
        self.lstm    = nn.LSTM(emb_dim, hidden_dim, n_layers, batch_first=True,
                               bidirectional=True,
                               dropout=dropout if n_layers > 1 else 0)
        self.fc      = nn.Linear(hidden_dim * 2, 1)   # forward + backward
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        emb = self.dropout(self.embedding(x))
        _, (hidden, _) = self.lstm(emb)
        # 마지막 레이어의 forward / backward hidden 연결
        hidden = torch.cat((hidden[-2], hidden[-1]), dim=1)
        return self.fc(self.dropout(hidden)).squeeze(1)


# ──────────────────────────────────────────────────────────────
# 모델 4: TextCNN (다중 커널 크기 병렬 컨볼루션)
# ──────────────────────────────────────────────────────────────
class TextCNNModel(nn.Module):
    def __init__(self, vocab_size, emb_dim, n_filters, filter_sizes, dropout,
                 pretrained_emb=None, freeze_emb=False):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        if pretrained_emb is not None:
            self.embedding.weight.data.copy_(pretrained_emb)
        if freeze_emb:
            self.embedding.weight.requires_grad = False
        self.convs = nn.ModuleList([
            nn.Conv2d(1, n_filters, (fs, emb_dim)) for fs in filter_sizes
        ])
        self.fc      = nn.Linear(n_filters * len(filter_sizes), 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        emb    = self.embedding(x).unsqueeze(1)             # (B, 1, L, E)
        pooled = [torch.relu(conv(emb)).squeeze(3).max(dim=2).values
                  for conv in self.convs]                   # 각 (B, F)
        cat    = self.dropout(torch.cat(pooled, dim=1))     # (B, F*num_sizes)
        return self.fc(cat).squeeze(1)


print('4가지 모델 클래스 정의 완료')
```

## 7. 모델 훈련 개시

```python
def train_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss, total_acc, n = 0.0, 0, 0
    for X_batch, y_batch in loader:
        X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
        optimizer.zero_grad()
        logits = model(X_batch)
        loss   = criterion(logits, y_batch)
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        preds       = (torch.sigmoid(logits) >= 0.5).float()
        total_loss += loss.item() * len(y_batch)
        total_acc  += (preds == y_batch).sum().item()
        n          += len(y_batch)
    return total_loss / n, total_acc / n


def evaluate(model, loader, criterion):
    model.eval()
    total_loss, total_acc, n = 0.0, 0, 0
    with torch.no_grad():
        for X_batch, y_batch in loader:
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            logits = model(X_batch)
            loss   = criterion(logits, y_batch)
            preds  = (torch.sigmoid(logits) >= 0.5).float()
            total_loss += loss.item() * len(y_batch)
            total_acc  += (preds == y_batch).sum().item()
            n          += len(y_batch)
    return total_loss / n, total_acc / n


def run_experiment(model, model_name, epochs=EPOCHS, lr=1e-3):
    """모델을 훈련하고 최적 가중치 기준 테스트 정확도를 반환합니다."""
    model     = model.to(DEVICE)
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)
    criterion = nn.BCEWithLogitsLoss()
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=2, factor=0.5)

    history      = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_acc = 0.0
    save_path    = f'/tmp/{model_name}_best.pt'

    params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f'\n{"="*58}')
    print(f'  Model : {model_name}   (Params: {params:,})')
    print(f'{"="*58}')
    print(f'  {"Epoch":>5} | {"Tr Loss":>8} | {"Tr Acc":>7} | {"Val Loss":>9} | {"Val Acc":>8}')
    print(f'  {"-"*52}')

    for epoch in range(1, epochs + 1):
        tr_loss, tr_acc = train_epoch(model, train_loader, optimizer, criterion)
        vl_loss, vl_acc = evaluate(model, val_loader, criterion)
        scheduler.step(vl_loss)

        history['train_loss'].append(tr_loss)
        history['val_loss'].append(vl_loss)
        history['train_acc'].append(tr_acc)
        history['val_acc'].append(vl_acc)

        if vl_acc > best_val_acc:
            best_val_acc = vl_acc
            torch.save(model.state_dict(), save_path)

        print(f'  {epoch:>5} | {tr_loss:>8.4f} | {tr_acc:>6.2%} | {vl_loss:>9.4f} | {vl_acc:>7.2%}')

    # 최적 가중치로 복원 후 테스트
    model.load_state_dict(torch.load(save_path))
    te_loss, te_acc = evaluate(model, test_loader, criterion)
    print(f'\n  Test Accuracy : {te_acc:.4f}  ({te_acc:.2%})')
    history['test_acc'] = te_acc
    return model, history


print('훈련 함수 정의 완료')
```

```python
# 실험 1: SimpleRNN
rnn_model = SimpleRNNModel(VOCAB_SIZE, EMB_DIM, HIDDEN_DIM, N_LAYERS, DROPOUT)
rnn_model, rnn_hist = run_experiment(rnn_model, 'SimpleRNN')
```

```python
# 실험 2: LSTM
lstm_model = LSTMModel(VOCAB_SIZE, EMB_DIM, HIDDEN_DIM, N_LAYERS, DROPOUT)
lstm_model, lstm_hist = run_experiment(lstm_model, 'LSTM')
```

```python
# 실험 3: Bidirectional LSTM
bilstm_model = BiLSTMModel(VOCAB_SIZE, EMB_DIM, HIDDEN_DIM, N_LAYERS, DROPOUT)
bilstm_model, bilstm_hist = run_experiment(bilstm_model, 'BiLSTM')
```

```python
# 실험 4: TextCNN
cnn_model = TextCNNModel(VOCAB_SIZE, EMB_DIM,
                         n_filters=128, filter_sizes=[2, 3, 4, 5], dropout=DROPOUT)
cnn_model, cnn_hist = run_experiment(cnn_model, 'TextCNN')
```

## 8. Loss & Accuracy 그래프 시각화

```python
all_models = {
    'SimpleRNN' : rnn_hist,
    'LSTM'      : lstm_hist,
    'BiLSTM'    : bilstm_hist,
    'TextCNN'   : cnn_hist,
}
colors       = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
epochs_range = range(1, EPOCHS + 1)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
for (name, hist), color in zip(all_models.items(), colors):
    axes[0].plot(epochs_range, hist['train_loss'], '--', color=color, alpha=0.5)
    axes[0].plot(epochs_range, hist['val_loss'],   '-',  color=color, label=name)
    axes[1].plot(epochs_range, hist['train_acc'],  '--', color=color, alpha=0.5)
    axes[1].plot(epochs_range, hist['val_acc'],    '-',  color=color, label=name)

for ax, title, ylabel in zip(
    axes,
    ['Loss 비교 (실선=Val, 점선=Train)', 'Accuracy 비교 (실선=Val, 점선=Train)'],
    ['BCE Loss', 'Accuracy']
):
    ax.set_title(title, fontsize=13)
    ax.set_xlabel('Epoch')
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print('\n=== 모델별 Test Accuracy 요약 ===')
for name, hist in all_models.items():
    print(f'  {name:<12}: {hist["test_acc"]:.2%}')
```

## 9. 학습된 Embedding 레이어 분석 (gensim 유사단어 찾기)

```python
from gensim.models import Word2Vec

# X_train 인덱스 시퀀스를 단어 문자열 시퀀스로 복원
train_word_seqs = [
    [index_to_word.get(idx, '<UNK>') for idx in seq if idx != 0]
    for seq in X_train
]

print('자체 Word2Vec 학습 중...')
w2v_self = Word2Vec(
    sentences   = train_word_seqs,
    vector_size = 128,
    window      = 5,
    min_count   = 2,
    workers     = 4,
    epochs      = 10
)
print(f'자체 학습 완료 — 사전 크기: {len(w2v_self.wv):,}')
```

```python
# 자체 학습 임베딩: 유사 단어 확인
test_words = ['재미', '연기', '최악', '감동', '스토리']

print('=== 자체 학습 Word2Vec 유사 단어 ===\n')
for word in test_words:
    if word in w2v_self.wv:
        sims = w2v_self.wv.most_similar(word, topn=5)
        print(f'  [{word}] -> ' + ', '.join(f'{w}({s:.3f})' for w, s in sims))
    else:
        print(f'  [{word}] -> 사전 외 단어')
```

```python
# 사전 학습 Word2Vec 로드 (gensim .load() 방식)
WORD2VEC_PATH = '/data/word2vec_ko.model'   # 실제 경로에 맞게 수정

try:
    w2v_pretrained = Word2Vec.load(WORD2VEC_PATH)
    pretrained_wv  = w2v_pretrained.wv          # .wv 로 KeyedVectors 접근
    print(f'사전 학습 모델 로드 완료 — 사전 크기: {len(pretrained_wv):,}')
    print(f'임베딩 차원: {pretrained_wv.vector_size}')
    PRETRAINED_OK = True
except Exception as e:
    print(f'사전 학습 모델 로드 실패: {e}')
    print('자체 학습 임베딩으로 대체합니다.')
    pretrained_wv = w2v_self.wv
    PRETRAINED_OK = False
```

```python
# 자체 학습 vs 사전 학습 임베딩 유사 단어 비교
print('=== 자체 학습 vs 사전 학습 임베딩 비교 분석 ===\n')
compare_words = ['재미', '연기', '최악', '감동', '스토리']

for word in compare_words:
    self_sim = ([f'{w}({s:.3f})' for w, s in w2v_self.wv.most_similar(word, topn=3)]
                if word in w2v_self.wv else ['없음'])
    pre_sim  = ([f'{w}({s:.3f})' for w, s in pretrained_wv.most_similar(word, topn=3)]
                if word in pretrained_wv else ['없음'])
    print(f'  [{word}]')
    print(f'    자체 학습  : {" | ".join(self_sim)}')
    print(f'    사전 학습  : {" | ".join(pre_sim)}')
    print()
```

## 10. 한국어 Word2Vec 임베딩 활용하여 성능 개선

사전 학습 임베딩을 임베딩 레이어에 주입(fine-tuning)하여 BiLSTM, TextCNN을 재학습합니다.

```python
def build_pretrained_embedding_matrix(word_to_index, wv, emb_dim):
    """
    word_to_index 사전 기준으로 사전 학습 임베딩 행렬을 구성합니다.
    사전에 없는 단어는 Xavier 균등 분포로 초기화합니다.
    """
    matrix = np.random.uniform(-0.1, 0.1, (len(word_to_index), emb_dim))
    matrix[0] = 0.0   # <PAD> = 0 벡터
    hit = miss = 0
    for word, idx in word_to_index.items():
        if word in wv:
            vec = wv[word]
            matrix[idx, :min(len(vec), emb_dim)] = vec[:emb_dim]
            hit += 1
        else:
            miss += 1
    print(f'Hit: {hit:,} / Miss: {miss:,}  (Hit율: {hit/(hit+miss):.2%})')
    return torch.tensor(matrix, dtype=torch.float)

pretrained_emb_dim = pretrained_wv.vector_size
print(f'사전 학습 임베딩 차원: {pretrained_emb_dim}')

pretrained_matrix = build_pretrained_embedding_matrix(
    word_to_index, pretrained_wv, pretrained_emb_dim
)
```

```python
# 실험 5: BiLSTM + 사전 학습 임베딩 (fine-tuning)
bilstm_pre = BiLSTMModel(
    VOCAB_SIZE, pretrained_emb_dim, HIDDEN_DIM, N_LAYERS, DROPOUT,
    pretrained_emb=pretrained_matrix, freeze_emb=False
)
bilstm_pre, bilstm_pre_hist = run_experiment(bilstm_pre, 'BiLSTM_Pretrained')
```

```python
# 실험 6: TextCNN + 사전 학습 임베딩 (fine-tuning)
cnn_pre = TextCNNModel(
    VOCAB_SIZE, pretrained_emb_dim,
    n_filters=128, filter_sizes=[2, 3, 4, 5], dropout=DROPOUT,
    pretrained_emb=pretrained_matrix, freeze_emb=False
)
cnn_pre, cnn_pre_hist = run_experiment(cnn_pre, 'TextCNN_Pretrained')
```

```python
# 전체 실험 결과 비교
all_results = {
    'SimpleRNN'          : rnn_hist,
    'LSTM'               : lstm_hist,
    'BiLSTM'             : bilstm_hist,
    'TextCNN'            : cnn_hist,
    'BiLSTM+Pretrained'  : bilstm_pre_hist,
    'TextCNN+Pretrained' : cnn_pre_hist,
}

print('\n' + '='*48)
print(f'  {"모델":<24} | {"Test Accuracy":>12}')
print('='*48)
for name, hist in all_results.items():
    acc  = hist['test_acc']
    flag = '  <- 목표 달성!' if acc >= 0.85 else ''
    print(f'  {name:<24} | {acc:>11.2%}{flag}')
print('='*48)
```

```python
# 최종 성능 비교 막대 그래프
names      = list(all_results.keys())
accs       = [all_results[n]['test_acc'] for n in names]
bar_colors = ['#4C72B0'] * 4 + ['#DD8452'] * 2

fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.barh(names, accs, color=bar_colors, edgecolor='white', height=0.55)

for bar, acc in zip(bars, accs):
    ax.text(bar.get_width() + 0.002, bar.get_y() + bar.get_height() / 2,
            f'{acc:.2%}', va='center', fontsize=11)

ax.axvline(0.85, color='red', linestyle='--', linewidth=1.5, label='목표: 85%')
ax.set_xlim(0.70, 0.92)
ax.set_xlabel('Test Accuracy')
ax.set_title('모델별 Test Accuracy 최종 비교\n(파랑=랜덤 임베딩 / 주황=사전 학습 임베딩)', fontsize=13)
ax.legend()
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 11. 실제 리뷰 감성 예측 테스트

```python
def predict_sentiment(sentence, model, word_to_index, max_len=MAX_LEN):
    model.eval()
    tokens  = tokenizer.morphs(sentence)
    tokens  = [w for w in tokens if w not in stopwords and len(w) > 1]
    encoded = [word_to_index.get(w, word_to_index['<UNK>']) for w in tokens]
    encoded = encoded[:max_len] + [0] * max(0, max_len - len(encoded))
    x = torch.tensor([encoded], dtype=torch.long).to(DEVICE)
    with torch.no_grad():
        prob = torch.sigmoid(model(x)).item()
    label = '긍정' if prob >= 0.5 else '부정'
    return label, prob

# 가장 높은 성능의 모델 선택
best_name = max(all_results, key=lambda k: all_results[k]['test_acc'])
model_map = {
    'SimpleRNN'          : rnn_model,
    'LSTM'               : lstm_model,
    'BiLSTM'             : bilstm_model,
    'TextCNN'            : cnn_model,
    'BiLSTM+Pretrained'  : bilstm_pre,
    'TextCNN+Pretrained' : cnn_pre,
}
best_model = model_map[best_name]

sample_reviews = [
    '연출이 너무 지루하고 배우들 연기도 엉망이었어요. 돈 아까워.',
    '정말 감동적인 영화였습니다. 눈물이 멈추지 않았어요.',
    '스토리가 허술하고 중간에 잠들 뻔 했습니다.',
    '배우들의 연기가 훌륭하고 영상미도 아름다운 걸작입니다.',
    '이런 영화를 극장에서 볼 줄이야... 완전 망작.',
    '가족과 함께 보기 좋은 따뜻한 영화. 강력 추천!'
]

print(f'사용 모델: {best_name}\n')
print(f'{"리뷰":<38} | 예측  | 확률')
print('-' * 58)
for review in sample_reviews:
    label, prob = predict_sentiment(review, best_model, word_to_index)
    print(f'{review[:37]:<38} | {label} | {prob:.4f}')
```
